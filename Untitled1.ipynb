```python
#!/usr/bin/env python
# coding: utf-8

"""
S&P 500 NEXT-5-DAY FORECAST MODEL
Multi-model training with metadata saving
"""

import numpy as np
import pandas as pd
import os
import pickle
from datetime import datetime, timedelta
from pandas_datareader import data as pdr
import warnings
warnings.filterwarnings("ignore")

# ML
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error, r2_score
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.ensemble import RandomForestRegressor
from sklearn.multioutput import MultiOutputRegressor
import xgboost as xgb

# Deep Learning
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.optimizers import Adam

np.random.seed(42)

print("="*60)
print("TRAINING NEXT-5-DAY FORECAST MODELS")
print("="*60)

os.makedirs("models", exist_ok=True)

# =====================================================
# 1. DATA COLLECTION
# =====================================================

tickers = ['AAPL','MSFT','GOOGL','AMZN','NVDA','META','JPM']
end = datetime.now()
start = end - timedelta(days=5*365)

all_data = {}

for ticker in tickers:
    df = pdr.DataReader(ticker, 'stooq', start, end)
    df.sort_index(inplace=True)
    all_data[ticker] = df
    print(f"{ticker}: {len(df)} rows")

# =====================================================
# 2. PREPROCESSING
# =====================================================

def preprocess_data(df):
    df = df.copy()
    df.sort_index(inplace=True)
    df.fillna(method="ffill", inplace=True)

    df['Close_Lag1'] = df['Close'].shift(1)
    df['Close_Lag2'] = df['Close'].shift(2)
    df['Close_Lag3'] = df['Close'].shift(3)
    df['SMA_10'] = df['Close'].rolling(10).mean().shift(1)
    df['SMA_20'] = df['Close'].rolling(20).mean().shift(1)
    df['Volatility_10'] = df['Close'].pct_change().rolling(10).std().shift(1)

    # Multi-step targets
    for i in range(1,6):
        df[f'Target_t+{i}'] = df['Close'].shift(-i)

    df.dropna(inplace=True)
    return df

processed = {t: preprocess_data(df) for t,df in all_data.items()}

feature_columns = [
    'Close_Lag1','Close_Lag2','Close_Lag3',
    'SMA_10','SMA_20','Volatility_10'
]

target_columns = [f'Target_t+{i}' for i in range(1,6)]

X_list, y_list = [], []

for df in processed.values():
    X_list.append(df[feature_columns].values)
    y_list.append(df[target_columns].values)

X = np.vstack(X_list)
y = np.vstack(y_list)

# =====================================================
# 3. TRAIN TEST SPLIT
# =====================================================

split = int(len(X)*0.8)
X_train, X_test = X[:split], X[split:]
y_train, y_test = y[:split], y[split:]

# =====================================================
# 4. SCALE (TRAIN ONLY)
# =====================================================

scaler_X = MinMaxScaler()
scaler_y = MinMaxScaler()

X_train_scaled = scaler_X.fit_transform(X_train)
X_test_scaled = scaler_X.transform(X_test)

y_train_scaled = scaler_y.fit_transform(y_train)

results = {}

# =====================================================
# 5. TRAIN MODELS
# =====================================================

def evaluate_model(name, model):
    model.fit(X_train_scaled, y_train_scaled)
    pred_scaled = model.predict(X_test_scaled)
    pred = scaler_y.inverse_transform(pred_scaled)

    mae = mean_absolute_error(y_test, pred)
    r2 = r2_score(y_test, pred)

    results[name] = {
        "model": model,
        "mae": mae,
        "r2": r2
    }

    print(f"{name} -> MAE: {mae:.4f} | R虏: {r2:.4f}")

# 1. Linear Regression
evaluate_model(
    "Linear Regression",
    MultiOutputRegressor(LinearRegression())
)

# 2. Ridge
evaluate_model(
    "Ridge",
    MultiOutputRegressor(Ridge(alpha=1.0))
)

# 3. Random Forest
evaluate_model(
    "Random Forest",
    MultiOutputRegressor(
        RandomForestRegressor(
            n_estimators=200,
            max_depth=10,
            random_state=42,
            n_jobs=-1
        )
    )
)

# 4. XGBoost
evaluate_model(
    "XGBoost",
    MultiOutputRegressor(
        xgb.XGBRegressor(
            n_estimators=200,
            max_depth=5,
            learning_rate=0.05,
            random_state=42
        )
    )
)

# 5. LSTM
print("Training LSTM...")

timesteps = 10
X_lstm, y_lstm = [], []

for i in range(len(X_train_scaled)-timesteps):
    X_lstm.append(X_train_scaled[i:i+timesteps])
    y_lstm.append(y_train_scaled[i+timesteps])

X_lstm = np.array(X_lstm)
y_lstm = np.array(y_lstm)

lstm_model = Sequential([
    LSTM(64, input_shape=(timesteps, X_train_scaled.shape[1])),
    Dropout(0.2),
    Dense(32, activation='relu'),
    Dense(5)
])

lstm_model.compile(optimizer=Adam(0.001), loss='mse')

early_stop = EarlyStopping(patience=5, restore_best_weights=True)

lstm_model.fit(X_lstm, y_lstm,
               epochs=30,
               batch_size=32,
               callbacks=[early_stop],
               verbose=1)

# Evaluate LSTM
X_test_lstm = []
for i in range(len(X_test_scaled)-timesteps):
    X_test_lstm.append(X_test_scaled[i:i+timesteps])

X_test_lstm = np.array(X_test_lstm)

pred_scaled = lstm_model.predict(X_test_lstm)
pred = scaler_y.inverse_transform(pred_scaled)

y_test_cut = y_test[timesteps:]

mae = mean_absolute_error(y_test_cut, pred)
r2 = r2_score(y_test_cut, pred)

results["LSTM"] = {
    "model": lstm_model,
    "mae": mae,
    "r2": r2
}

print(f"LSTM -> MAE: {mae:.4f} | R虏: {r2:.4f}")

# =====================================================
# 6. SELECT BEST MODEL
# =====================================================

best_model_name = min(results, key=lambda x: results[x]["mae"])
best_model = results[best_model_name]["model"]
best_mae = results[best_model_name]["mae"]
best_r2 = results[best_model_name]["r2"]

print("\nBEST MODEL:", best_model_name)

# =====================================================
# 7. SAVE MODEL + METADATA (YOUR ORIGINAL STRUCTURE)
# =====================================================

if best_model_name == "LSTM":
    best_model.save("models/stock_predictor.h5")
    model_type = "lstm"
else:
    with open("models/stock_predictor.pkl","wb") as f:
        pickle.dump(best_model,f)
    model_type = "sklearn"

with open("models/scaler_X.pkl","wb") as f:
    pickle.dump(scaler_X,f)

with open("models/scaler_y.pkl","wb") as f:
    pickle.dump(scaler_y,f)

with open("models/features.pkl","wb") as f:
    pickle.dump(feature_columns,f)

metadata = {
    "best_model": best_model_name,
    "mae_score": best_mae,
    "r2_score": best_r2,
    "forecast_horizon": 5,
    "features": feature_columns,
    "model_type": model_type,
    "training_date": datetime.now().strftime("%Y-%m-%d %H:%M:%S"),
    "num_samples": len(X_train),
    "all_results": {
        name: {"mae": m["mae"], "r2": m["r2"]}
        for name, m in results.items()
    }
}

with open("models/metadata.pkl","wb") as f:
    pickle.dump(metadata,f)

print("\nModel and metadata saved successfully.")
```
